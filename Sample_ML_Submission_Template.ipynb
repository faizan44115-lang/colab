## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Copy dataset
df1 = df.copy()

# Remove unwanted column
df1.drop('Unnamed: 0', axis=1, inplace=True)

# Remove rating from company name
df1['Company Name'] = df1['Company Name'].apply(lambda x: x.split('\n')[0])

# Handle missing values
df1['Rating'] = df1['Rating'].fillna(df1['Rating'].median())

# Remove duplicate rows
df1 = df1.drop_duplicates()

# Clean Salary Estimate column
df1['Salary Estimate'] = df1['Salary Estimate'].str.replace('(Glassdoor est.)', '', regex=False)
df1['Salary Estimate'] = df1['Salary Estimate'].str.replace('$', '', regex=False)
df1['Salary Estimate'] = df1['Salary Estimate'].str.replace('K', '', regex=False)

# Split salary range
salary = df1['Salary Estimate'].str.split('-', expand=True)

# Create minimum and maximum salary columns
df1['Min Salary'] = pd.to_numeric(salary[0], errors='coerce')
df1['Max Salary'] = pd.to_numeric(salary[1], errors='coerce')

# Create average salary column
df1['Avg Salary'] = (df1['Min Salary'] + df1['Max Salary']) / 2

# Check cleaned dataset
df1.head()
````

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Null Hypothesis**

There is no significant relationship between company ratings and average salary offered for Data Science jobs.

**Alternate Hypothesis**

There is a significant relationship between company ratings and average salary offered for Data Science jobs.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import pearsonr

# Remove missing values
temp_df = df1[['Rating', 'Avg Salary']].dropna()

# Perform Pearson Correlation Test
corr, p_value = pearsonr(temp_df['Rating'], temp_df['Avg Salary'])

print("Correlation Coefficient:", corr)
print("P-Value:", p_value)
````

##### Which statistical test have you done to obtain P-Value?

The Pearson Correlation Test

##### Why did you choose the specific statistical test?

The Pearson Correlation Test was chosen because both Rating and Avg Salary are continuous numerical variables. The objective was to measure the strength and direction of the relationship between these two variables.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Null Hypothesis**

Different industries do not offer significantly different average salaries for Data Science jobs.

**Alternate Hypothesis**

Different industries offer significantly different average salaries for Data Science jobs.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import f_oneway

# Select top industries
top_industries = df['Industry'].value_counts().head(3).index

# Create salary groups
group1 = df1[df1['Industry'] == top_industries[0]]['Avg Salary'].dropna()
group2 = df1[df1['Industry'] == top_industries[1]]['Avg Salary'].dropna()
group3 = df1[df1['Industry'] == top_industries[2]]['Avg Salary'].dropna()

# Perform ANOVA Test
f_stat, p_value = f_oneway(group1, group2, group3)

print("F-Statistic:", f_stat)
print("P-Value:", p_value)
````

##### Which statistical test have you done to obtain P-Value?

The One-Way ANOVA (Analysis of Variance) test

##### Why did you choose the specific statistical test?

The One-Way ANOVA test was selected because the analysis involved comparing average salaries across more than two industry groups. ANOVA determines whether there is a statistically significant difference between the means of multiple groups.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Null Hypothesis**

Large companies do not offer higher average salaries compared to small companies.

**Alternate Hypothesis**

Large companies offer higher average salaries compared to small companies.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import ttest_ind

# Example company size groups
large_companies = df1[df1['Size'].isin(['10000+ employees'])]['Avg Salary'].dropna()

small_companies = df1[df1['Size'].isin(['1 to 50 employees'])]['Avg Salary'].dropna()

# Perform T-Test
t_stat, p_value = ttest_ind(large_companies, small_companies)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)
````

##### Which statistical test have you done to obtain P-Value?

The Independent T-Test

##### Why did you choose the specific statistical test?

The Independent T-Test was used because the comparison was between two independent groups: large companies and small companies.

````python
# Lower Casing
df1['Job Description'] = df1['Job Description'].str.lower()

df1['Job Description'].head()
````

#### 3. Removing Punctuations

````python
# Remove Punctuations
import string

df1['Job Description'] = df1['Job Description'].str.translate(
    str.maketrans('', '', string.punctuation)
)

df1['Job Description'].head()
````

#### 4. Removing URLs & Removing words and digits contain digits.

````python
# Remove URLs & Remove words and digits contain digits
# Remove URLs and digits

import re

def clean_text(text):

    # Remove URLs
    text = re.sub(r'http\\S+|www\\S+', '', text)

    # Remove digits
    text = re.sub(r'\\d+', '', text)

    return text

df1['Job Description'] = df1['Job Description'].apply(clean_text)

df1['Job Description'].head()
````

#### 5. Removing Stopwords & Removing White spaces

````python
# Download stopwords

import nltk

nltk.download('stopwords')
````

````python
# Remove Stopwords

from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def remove_stopwords(text):

    words = text.split()

    filtered_words = [word for word in words if word not in stop_words]

    return " ".join(filtered_words)

df1['Job Description'] = df1['Job Description'].apply(remove_stopwords)
````

````python
# Remove White spaces

df1['Job Description'] = df1['Job Description'].str.strip()

df1['Job Description'].head()
````

#### 6. Rephrase Text

````python
# Rephrase Text
df1['Job Description'] = df1['Job Description'].str.replace(
    'machine learning',
    'ml'
)

df1['Job Description'].head()
````

#### 7. Tokenization

````python
# Download punkt

nltk.download('punkt')
nltk.download('punkt_tab')
# Download wordnet

nltk.download('wordnet')
````

````python
# Tokenization
from nltk.tokenize import word_tokenize

df1['Tokens'] = df1['Job Description'].apply(word_tokenize)

df1[['Job Description', 'Tokens']].head()
````

#### 8. Text Normalization

````python
# Normalizing Text (i.e., Stemming, Lemmatization etc.)
# Stemming

from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

df1['Stemmed'] = df1['Tokens'].apply(
    lambda words: [stemmer.stem(word) for word in words]
)

df1[['Tokens', 'Stemmed']].head()
````
